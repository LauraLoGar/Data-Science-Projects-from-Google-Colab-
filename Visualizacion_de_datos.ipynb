```python
#relaciones estadisticas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
#primer scatter plot. vamos a utilizar relplot y pasaremos 3 inputs.
#importante poner ; al final para tener la info mejor.
tips = sns.load_dataset('tips')
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
#diferencias categoricas
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
#jugamos con el tipo de marcador, aqui lo cambiamos con style
sns.relplot(x='total_bill', y='tip', hue='smoker', style='time', data=tips);
```

```python
#mas relaciones numericas =. variable categoridca donde el orden importa.
sns.relplot(x='total_bill', y='tip', hue='size', data=tips);
```

```python
#cambiar la paleta de colores.
sns.relplot(x='total_bill', y='tip', hue='size',
            palette='ch:r=-.5,l=.75', data=tips);
```

```python
#modificar el campo size. a mas numero, mayor tamaño
sns.relplot(x='total_bill', y='tip', size='size', data=tips);
```

```python
#como no se ve muy bien, se añade un segundo campo, sizes (tupla). ahi se pone el tamaño de mas peque y del mas grande.
sns.relplot(x='total_bill', y='tip', size='size',
            sizes=(15,200), data=tips);
```

```python
#line plot. kind=line significa que quiero un plot de lineas.
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x='time', y='value', kind='line', data=df)
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.displot(penguins, x='flipper_length_mm');
```

```python
#utilizo hue para tener un color por cada especie.
sns.displot(penguins, x='flipper_length_mm', hue='species');
```

```python
#para normalizar los datos - stat
sns.displot(penguins, x='flipper_length_mm',
            hue='species', stat='density');
```

```python
sns.displot(penguins, x='flipper_length_mm',
            hue='species', stat='probability', common_norm=False);
```

```python
#Boxplots
sns.boxplot(x='flipper_length_mm',
            data = penguins);
```

```python
#kernel density stimation (curva similar al histograma pero sin rectangulos)
sns.displot(penguins)
```

```python
sns.displot(penguins, x='flipper_length_mm',
            hue='species', kind='kde')
```

```python
sns.displot(penguins, x='flipper_length_mm',
            hue='species', kind='kde', fill=True)
```
